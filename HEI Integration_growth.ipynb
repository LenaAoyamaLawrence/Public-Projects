# The script loops through the folders and generate a dataframe (table) with:
* All HEI names
* Number of accounts at each HEI at different time point


```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# calculate pct change of the uni growth

def uni_growth(series_new, series_og):
    pct = np.absolute(series_new - series_og) / series_og
    pct = round(pct, 2)
    return pct
```

## Drupal merge

```python
# import january 26
path_janD = 'C:/Users/ywa/Documents/Data/UserReport/Jan_2026_Moodle/hei_data_drupal.xlsx'

jan26_D = pd.read_excel(path_janD)


# import december 25
path_decD = 'C:/Users/ywa/Documents/Data/UserReport/Dec_2025_Moodle/hei_data_drupal.xlsx'

dec25_D = pd.read_excel(path_decD, usecols = ['count', 'provider'])

# merge jan26 and dec25

merged_drupal = jan26_D.merge(dec25_D, how = 'left', on = 'provider', validate = 'one_to_one')

merged_drupal.columns = ['provider', 'Jan_2026', 'state', 'Dec_2025']

# import september 25

path_sepD = 'C:/Users/ywa/Documents/Data/UserReport/Sep_2025_Moodle/hei_data_drupal.xlsx'

sep_25_D = pd.read_excel(path_sepD, usecols = ['count', 'provider'])

merged_drupal = merged_drupal.merge(sep_25_D, how = 'left', on = 'provider', validate = 'one_to_one')

merged_drupal.columns = ['provider', 'Jan_2026', 'state', 'Dec_2025', 'Sep_2025']

merged_drupal.head()

# import july 25

path_julD = 'C:/Users/ywa/Documents/Data/UserReport/Jul_2025_Moodle/hei_data_drupal.xlsx'

jul_25_D = pd.read_excel(path_julD, usecols = ['count', 'provider'])

merged_drupal = merged_drupal.merge(jul_25_D, how = 'left', on = 'provider', validate = 'one_to_one')

merged_drupal.columns = ['provider', 'Jan_2026', 'state', 'Dec_2025', 'Sep_2025', 'Jul_2025']

merged_drupal.head()

# import april 25

path_aprD = 'C:/Users/ywa/Documents/Data/UserReport/Apr_2025_Moodle/hei_data_drupal.xlsx'

apr_25_D = pd.read_excel(path_aprD, usecols = ['count', 'provider'])

merged_drupal = merged_drupal.merge(apr_25_D, how = 'left', on = 'provider', validate = 'one_to_one')

merged_drupal.columns = ['provider', 'Jan_2026', 'state', 'Dec_2025', 'Sep_2025', 'Jul_2025', 'Apr_2025']

# reordering columns

merged_drupal = merged_drupal.iloc[:, [0, 2, 1, 3, 4, 5, 6]]

merged_drupal.head()
```

```python
# deal with missing values (due to merge)
merged_drupal.fillna(0, inplace = True)
```

```python


merged_drupal['rate_JanDec2526'] = uni_growth(merged_drupal['Jan_2026'], merged_drupal['Dec_2025'])
merged_drupal['rate_DecSep25'] = uni_growth(merged_drupal['Dec_2025'], merged_drupal['Sep_2025'])
merged_drupal['rate_SepJul25'] = uni_growth(merged_drupal['Sep_2025'], merged_drupal['Jul_2025'])
merged_drupal['rate_JulApr25'] = uni_growth(merged_drupal['Jul_2025'], merged_drupal['Apr_2025'])

#merged_drupal.head()

#merged_drupal = merged_drupal.iloc[:, [0, 1, 2, 7, 3, 8, 4, 9, 5, 10, 6]]

merged_drupal.head(10)
```
